# Exploratory Data Analysis (EDA)

In this section I explore the dataset to understand class balance, key distributions,
and early relationships with the target (`readmitted`). All visuals are followed by brief,
business-focused takeaways.

```python
import pandas as pd

# Load the dataset
file_path = "data/diabetes/diabetic_data.csv"
df = pd.read_csv(file_path)

# Step 1 checks
print("Dataset shape:", df.shape)
print("\nColumns and types:\n", df.dtypes)
print("\nFirst 5 rows:\n", df.head())
print("\nMissing values per column:\n", df.isnull().sum())
```

```python
# Step 2a:Drop duplicate rows
df = df.drop_duplicates()
print("Shape after dropping duplicates:", df.shape)

# Step 2b: Replace '?' with NaN in object columns
df.replace('?', pd.NA, inplace=True)

# Check missing values again
missing_values = df.isnull().sum()
missing_values[missing_values > 0]
```

```python
# Drop columns with very high missing rate (~>80% missing):
df.drop(columns=['weight', 'max_glu_serum', 'A1Cresult'], inplace=True)
```

```python
# Step 3: Identify categorical and numerical columns
categorical_cols = df.select_dtypes(include='object').columns.tolist()
numerical_cols   = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
```

```python
categorical_cols
```

```python
# Step 4: Fill missing values based on type
# For categorical columns: fill with 'Unknown'
for col in categorical_cols:
    df[col] = df[col].fillna('Unknown')

# For numerical columns: fill with median
for col in numerical_cols:
    df[col] = df[col].fillna(df[col].median())

# Check missing values after imputation
missing_after = df.isnull().sum()
missing_after[missing_after > 0]
```

### Exploratory Data Analysis (EDA Overview)

```python
# Step 5: Exploratory Data Analysis (EDA Overview)

# Basic statistics for numerical features
print("=== Numerical Features Summary ===")
print(df[numerical_cols].describe().T)

# Count unique values for categorical features
print("\n=== Categorical Features Unique Values ===")
for col in categorical_cols:
    unique_vals = df[col].nunique()
    print(f"{col}: {unique_vals} unique values")

# Check target variable distribution
if 'readmitted' in df.columns:
    print("\n=== Target Variable ('readmitted') Distribution ===")
    print(df['readmitted'].value_counts())
else:
    print("\n 'readmitted' column not found â€” please confirm target variable name.")
```

```python
print("Any remaining '?' values:", (df == '?').sum().sum())
print("Any remaining NaN values:", df.isna().sum().sum())
```

### Goal.
Understand class balance of `readmitted` to anticipate modeling challenges.

Expectation.
Class imbalance is common in readmission tasks; weâ€™ll verify the ratio of:
   - `NO` (no readmission)
   - `>30` (readmitted after 30 days)
   - `<30` (readmitted within 30 days â€” often the primary business focus)

```python
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Make sure the folder exists before saving
os.makedirs("plots", exist_ok=True)

# Create the figure
plt.figure(figsize=(6,4))

# Fixed: add hue='readmitted' and legend=False to remove FutureWarning
ax = sns.countplot(
    data=df,
    x='readmitted',
    hue='readmitted',          # required by new Seaborn versions
    order=['NO','>30','<30'],
    palette='Set2',
    legend=False
)

# Add percentage labels on top of each bar
total = len(df)
for p in ax.patches:
    percentage = f'{100 * p.get_height() / total:.1f}%'
    x = p.get_x() + p.get_width() / 2
    y = p.get_height()
    ax.text(x, y + 500, percentage, ha='center', fontsize=11)

# Titles and labels
plt.title("Readmission Status Distribution")
plt.xlabel("Readmission Category")
plt.ylabel("Number of Patients")

# Save the figure BEFORE showing it
plt.savefig("plots/readmission_status_distrib.png", dpi=300, bbox_inches='tight')

# Display the chart
plt.show()

print("Saved chart as: plots/readmission_status_distrib.png")
```

```python
# Calculate percentage distribution for the 'readmitted' column
readmit_counts = df['readmitted'].value_counts()
readmit_percent = round((readmit_counts / len(df)) * 100, 2)

# Display both counts and percentages
display(pd.DataFrame({
    'Count': readmit_counts,
    'Percent': readmit_percent
}))
```

### Interpretation
  The majority class is NO (54%), with >30 (35%) and <30 (11%).
  This is a moderately imbalanced problem.  
  For modeling, Iâ€™ll consider stratified splits and metrics beyond accuracy
  (e.g., ROC-AUC, PR-AUC, F1 for `<30`), plus techniques like class weights.

## Distribution of Key Numeric Features

Goal:
Explore how core numeric features (hospital stay length, lab work, medication count, and diagnoses) are distributed among patients.

**Why it matters:**  
Understanding numeric feature distributions helps identify outliers, skewed variables, and general patterns of patient complexity.  
For instance, variables like `num_medications` or `num_lab_procedures` often reflect illness severity and treatment intensity â€” both could be predictive of readmission risk.

```python
numeric_features = ['time_in_hospital', 'num_lab_procedures', 
                    'num_medications', 'number_diagnoses']

fig, axes = plt.subplots(2, 2, figsize=(10,8))
axes = axes.flatten()

for i, col in enumerate(numeric_features):
    sns.histplot(df[col], bins=20, kde=True, ax=axes[i], color='skyblue', edgecolor='black')
    axes[i].set_title(f"{col} Distribution")
    axes[i].set_xlabel(col)
    axes[i].set_ylabel("Frequency")

plt.suptitle("Distribution of Key Numeric Features", fontsize=14)
plt.tight_layout()
plt.savefig("plots/distribution_features.png", dpi=300, bbox_inches='tight')
plt.show()
```

### Interpretation

- **`time_in_hospital`** is right-skewed â€” most patients stay between 1â€“7 days, with a small group having long hospitalizations.
- **`num_lab_procedures`** varies widely, suggesting diverse diagnostic intensity; some patients undergo over 100 procedures.
- **`num_medications`** also shows a long tail, indicating that certain patients require extensive medication management.
- **`number_diagnoses`** clusters around 6â€“9, consistent with chronic multi-diagnosis profiles.

**Insight:**  
These variables reflect different dimensions of patient complexity.  
For modeling, they might benefit from:
- **Log transformation** (for right-skewed features)
- **Normalization or scaling** before feeding into distance-based algorithms (e.g., Logistic Regression, SVM, or KNN).

## Average Hospital Stay by Readmission Category

**Goal:** Examine whether patients who are readmitted (especially within 30 days) tend to have longer hospital stays.

**Why it matters:**  
Length of stay can indicate illness severity, complications, or quality of care.  
If early readmissions show longer average stays, that may suggest these patients are medically complex and high risk for returning.

```python
plt.figure(figsize=(6,4))
ax = sns.barplot(
    data=df,
    x='readmitted',
    y='time_in_hospital',
    order=['NO','>30','<30'],
    palette='coolwarm',
    estimator='mean'
)

# Add value labels on top of bars
for p in ax.patches:
    ax.annotate(
        f"{p.get_height():.2f}",
        (p.get_x() + p.get_width()/2., p.get_height()),
        ha='center', va='bottom', fontsize=11, color='black'
    )

plt.title("Average Hospital Stay by Readmission Category")
plt.xlabel("Readmission Status")
plt.ylabel("Average Days in Hospital")
plt.tight_layout()
plt.savefig("plots/avg_hospital_stay_readmission_category.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
# Calculate the average (mean) hospital stay for each readmission category
avg_stay = df.groupby('readmitted')['time_in_hospital'].mean().round(2)
print(avg_stay)
```

```python
df.groupby('readmitted')['time_in_hospital'].agg(['mean','median','std','min','max']).round(2)
```

### Interpretation
Patients readmitted **within 30 days (`<30`)** show the **longest average hospital stay**.  
Those readmitted **after 30 days (`>30`)** have slightly shorter stays, while **non-readmitted (`NO`)** patients stay the fewest days on average.  
This pattern suggests a **positive relationship between hospital-stay length and early readmission**, consistent with more complex or severe clinical cases.  
`time_in_hospital` will likely be an important predictor; it may also interact with other features such as number of diagnoses or medications.

## Correlation Heatmap (Numeric Relationships)

Goal:
 Explore relationships between numeric features to understand how variables move together.

**Why it matters:**  
- Helps detect **multicollinearity** â€” when features are highly correlated, which can affect some models (like linear regression).  
- Reveals **feature clusters** â€” for example, if `num_lab_procedures` and `num_medications` tend to increase together, that could indicate more complex or severe patient cases.  
- Identifies which numeric variables might have **predictive strength** or redundancy before modeling.


```python
# Select only numeric columns for correlation
numeric_df = df.select_dtypes(include=['int64', 'float64'])

# Compute correlation matrix
corr = numeric_df.corr()

# Create heatmap
plt.figure(figsize=(12,8))
sns.heatmap(corr, cmap='coolwarm', center=0, annot=False, fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap of Numeric Features", fontsize=14, pad=15)
plt.savefig("plots/correlation_heatmap.png", dpi=300, bbox_inches='tight')
plt.show()
```

### Interpretation

- **Moderate correlations** are expected among hospital utilization features â€” for example,  
  `num_medications`, `num_lab_procedures`, and `number_diagnoses` may rise together as patient complexity increases.  
- **`time_in_hospital`** typically shows a mild positive correlation with those same features, confirming that longer stays often involve more tests and treatments.  
- There are **no extremely high correlations (r > 0.9)**, suggesting that multicollinearity is not a major concern at this stage.  
- These insights justify keeping all numeric variables for modeling, though we may apply **feature scaling** later.

**Insight for Modeling:**  
Since no features are redundant, itâ€™s safe to retain all numeric variables.  
However, models sensitive to scaling (e.g., Logistic Regression, SVM) should normalize these features to prevent bias from magnitude differences.

```python
# Sort and display top correlations (excluding self-correlation)
corr_pairs = corr.unstack().sort_values(ascending=False)
corr_pairs = corr_pairs[corr_pairs < 1]  # remove self-correlation (1.0)
print(corr_pairs.head(10))
```

### Interpretation (Detailed)

The top correlated pairs among numeric variables are:

| Feature 1 | Feature 2 | Correlation |
|------------|------------|-------------|
| encounter_id | patient_nbr | 0.51 |
| time_in_hospital | num_medications | 0.47 |
| num_medications | num_procedures | 0.39 |
| time_in_hospital | num_lab_procedures | 0.32 |
| num_lab_procedures | num_medications | 0.27 |

**Insights:**
- The correlation between **`encounter_id`** and **`patient_nbr`** (0.51) reflects patient-level grouping, not clinical meaning â€” both are identifiers.  
  These should be **dropped before modeling** to avoid leakage.  
- **`time_in_hospital`**, **`num_medications`**, and **`num_lab_procedures`** show **moderate positive correlations**, which makes clinical sense:  
  longer stays â†’ more tests â†’ more medications.  
- No pair exceeds 0.9, meaning **multicollinearity is not a serious issue**.  
- These results validate keeping all numeric variables for now, though weâ€™ll standardize them later for ML models.

**Modeling Implication:**  
Weâ€™ll remove identifier columns (`encounter_id`, `patient_nbr`) but retain all other numeric features.  
Normalization or scaling will be applied to ensure fair weight distribution across numeric variables.

## Readmission Rates by Age Group or Race

**Goal:**  
Understand whether certain age brackets or racial groups have higher readmission rates.

**Why it matters:**  
Demographic differences in hospital readmissions can reveal disparities or patterns in care outcomes.
This step helps identify whether age or race influences the likelihood of being readmitted.

```python
# Calculate percentage distribution by age and readmission category
age_readmit = (
    df.groupby('age')['readmitted']
    .value_counts(normalize=True)
    .unstack(fill_value=0)
    .sort_index()
) * 100

# Plot stacked bar chart
age_readmit[['NO', '>30', '<30']].plot(
    kind='bar', stacked=True, figsize=(10,5),
    color=['#7fc97f','#beaed4','#fdc086'], edgecolor='black'
)
plt.title("Readmission Percentage by Age Group")
plt.ylabel("Percentage of Patients")
plt.xlabel("Age Bracket")
plt.legend(title='Readmission Status')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("plots/readmission_by_age.png", dpi=300, bbox_inches='tight')
plt.show()
```

### Interpretation

- **Younger age brackets (30â€“40 years)** tend to have slightly higher proportions of `<30` readmissions.  
- **Older patients (70â€“90 years)** show more `>30` and `NO` readmissions, possibly reflecting stable long-term management.  
- No extreme disparities appear, but age clearly influences the readmission pattern.

**Insight:**  
Age will likely serve as an important categorical predictor in modeling.  
Later, weâ€™ll encode it ordinally (e.g., 0â€“9) to reflect increasing risk.

## Top 10 Diagnoses by Readmission

**Goal:**  
Identify which diagnoses (primary diagnosis `diag_1`) are most associated with early readmissions (`<30`).

**Why it matters:**  
Understanding which medical conditions drive early readmissions helps highlight high-risk patient groups
and may guide interventions to reduce hospital returns.

```python

# Step 1: compute the top 10 primary diagnoses
top_diag = (
    df[df['readmitted'] == '<30']['diag_1']
    .value_counts()
    .reset_index()
    .rename(columns={'index': 'Diagnosis_Code', 'diag_1': 'Patient_Count'})
    .head(10)
)

# Step 2: check columns
print("Columns in top_diag:", top_diag.columns)
display(top_diag.head())

# Step 3: handle alternative column names if Pandas renamed them differently
if 'count' in top_diag.columns:
    top_diag.rename(columns={'count': 'Patient_Count', 'Patient_Count': 'Diagnosis_Code'}, inplace=True)

# Step 4: make sure final column names are correct
print("Renamed columns:", top_diag.columns)

# Step 5: plot
plt.figure(figsize=(8,4))
sns.barplot(
    data=top_diag,
    x='Patient_Count',
    y='Diagnosis_Code',
    color='skyblue',
    edgecolor='black'
)

# Add labels
for i, row in top_diag.iterrows():
    plt.text(row['Patient_Count'] + 30, i, str(row['Patient_Count']), va='center')

plt.title("Top 10 Primary Diagnoses for <30-Day Readmissions", fontsize=13)
plt.xlabel("Number of Patients")
plt.ylabel("Diagnosis Code")
plt.tight_layout()
plt.savefig("plots/top_ten_diagnoses.png", dpi=300, bbox_inches='tight')
plt.show()

print("Saved chart as: plots/top_ten_diagnoses.png")
```
